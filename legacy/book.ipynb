# Practical Task A: Create VPC Components in Python
This section creates a complete VPC infrastructure with:
- VPC with CIDR block
- Public and Private Subnets
- Internet Gateway
- Route Tables and associations
- All resource IDs are printed for verification

```python
import boto3
import json

class VPCInfrastructure:
    """class to manage vpc infrastructure creation"""
    
    def __init__(self, region='us-east-1'):
        self.ec2_client = boto3.client('ec2', region_name=region)
        self.ec2_resource = boto3.resource('ec2', region_name=region)
        self.vpc_id = None
        self.public_subnet_id = None
        self.private_subnet_id = None
        self.igw_id = None
        self.route_table_id = None
        
    def create_vpc(self, cidr_block='10.0.0.0/16', vpc_name='MyVPC'):
        """create vpc with specified cidr block"""
        try:
            response = self.ec2_client.create_vpc(
                CidrBlock=cidr_block,
                TagSpecifications=[
                    {
                        'ResourceType': 'vpc',
                        'Tags': [{'Key': 'Name', 'Value': vpc_name}]
                    }
                ]
            )
            self.vpc_id = response['Vpc']['VpcId']
            print(f"vpc created: {self.vpc_id}")
            print(f"cidr block: {cidr_block}")
            return self.vpc_id
        except Exception as e:
            print(f"error creating vpc: {e}")
            return None
    
    def create_subnets(self):
        """create public and private subnets"""
        try:
            # create public subnet
            public_response = self.ec2_client.create_subnet(
                VpcId=self.vpc_id,
                CidrBlock='10.0.1.0/24',
                AvailabilityZone='us-east-1a',
                TagSpecifications=[
                    {
                        'ResourceType': 'subnet',
                        'Tags': [{'Key': 'Name', 'Value': 'PublicSubnet'}]
                    }
                ]
            )
            self.public_subnet_id = public_response['Subnet']['SubnetId']
            print(f"public subnet created: {self.public_subnet_id}")
            print(f"cidr block: 10.0.1.0/24")
            
            # create private subnet
            private_response = self.ec2_client.create_subnet(
                VpcId=self.vpc_id,
                CidrBlock='10.0.2.0/24',
                AvailabilityZone='us-east-1b',
                TagSpecifications=[
                    {
                        'ResourceType': 'subnet',
                        'Tags': [{'Key': 'Name', 'Value': 'PrivateSubnet'}]
                    }
                ]
            )
            self.private_subnet_id = private_response['Subnet']['SubnetId']
            print(f"private subnet created: {self.private_subnet_id}")
            print(f"cidr block: 10.0.2.0/24")
            
            return self.public_subnet_id, self.private_subnet_id
        except Exception as e:
            print(f"error creating subnets: {e}")
            return None, None
    
    def create_internet_gateway(self):
        """create and attach internet gateway to vpc"""
        try:
            # create internet gateway
            igw_response = self.ec2_client.create_internet_gateway(
                TagSpecifications=[
                    {
                        'ResourceType': 'internet-gateway',
                        'Tags': [{'Key': 'Name', 'Value': 'MyInternetGateway'}]
                    }
                ]
            )
            self.igw_id = igw_response['InternetGateway']['InternetGatewayId']
            print(f"internet gateway created: {self.igw_id}")
            
            # attach to vpc
            self.ec2_client.attach_internet_gateway(
                InternetGatewayId=self.igw_id,
                VpcId=self.vpc_id
            )
            print(f"internet gateway attached to vpc: {self.vpc_id}")
            
            return self.igw_id
        except Exception as e:
            print(f"error creating/attaching internet gateway: {e}")
            return None
    
    def create_route_table(self):
        """create route table and associate with public subnet"""
        try:
            # create route table
            rt_response = self.ec2_client.create_route_table(
                VpcId=self.vpc_id,
                TagSpecifications=[
                    {
                        'ResourceType': 'route-table',
                        'Tags': [{'Key': 'Name', 'Value': 'PublicRouteTable'}]
                    }
                ]
            )
            self.route_table_id = rt_response['RouteTable']['RouteTableId']
            print(f"route table created: {self.route_table_id}")
            
            # create route to internet gateway
            self.ec2_client.create_route(
                RouteTableId=self.route_table_id,
                DestinationCidrBlock='0.0.0.0/0',
                GatewayId=self.igw_id
            )
            print(f"route to internet gateway added (0.0.0.0/0 -> {self.igw_id})")
            
            # associate route table with public subnet
            association = self.ec2_client.associate_route_table(
                RouteTableId=self.route_table_id,
                SubnetId=self.public_subnet_id
            )
            print(f"route table associated with public subnet")
            
            return self.route_table_id
        except Exception as e:
            print(f"error creating/configuring route table: {e}")
            return None
    
    def create_full_infrastructure(self):
        """create complete vpc infrastructure"""
        print("creating vpc infrastructure")
        print()
        
        # create vpc
        if not self.create_vpc():
            return False
        
        print()
        
        # create subnets
        if not all(self.create_subnets()):
            return False
        
        print()
        
        # create and attach internet gateway
        if not self.create_internet_gateway():
            return False
        
        print()
        
        # create route table
        if not self.create_route_table():
            return False
        
        print()
        print("summary - resource ids for use in other modules:")
        print(f"vpc id: {self.vpc_id}")
        print(f"public subnet id: {self.public_subnet_id}")
        print(f"private subnet id: {self.private_subnet_id}")
        print(f"internet gateway: {self.igw_id}")
        print(f"route table id: {self.route_table_id}")
        
        return True
```

# Practical Task B: IAM Security Audit and Remediation
This section implements an automated security audit system that:
1. Scans all IAM users for over-privileged access
2. Identifies users with AdministratorAccess or wildcard permissions
3. Automatically detaches dangerous policies
4. Creates and attaches a safe LimitedAccessPolicy (S3 and EC2 ReadOnly)

```python
import boto3
import json
from typing import List, Dict

class IAMSecurityAuditor:
    """class to audit and remediate over-privileged iam users"""
    
    def __init__(self):
        self.iam_client = boto3.client('iam')
        self.flagged_users = []
        self.safe_policy_name = 'LimitedAccessPolicy'
        self.safe_policy_arn = None
        
    def get_all_users(self) -> List[str]:
        """retrieve all iam users in the account"""
        try:
            response = self.iam_client.list_users()
            users = [user['UserName'] for user in response['Users']]
            print(f"found {len(users)} iam users to audit")
            return users
        except Exception as e:
            print(f"error listing users: {e}")
            return []
    
    def check_managed_policies(self, username: str) -> List[Dict]:
        """check if user has dangerous managed policies attached"""
        dangerous_policies = []
        try:
            response = self.iam_client.list_attached_user_policies(UserName=username)
            
            for policy in response['AttachedPolicies']:
                policy_name = policy['PolicyName']
                policy_arn = policy['PolicyArn']
                
                # check for administratoraccess
                if policy_name == 'AdministratorAccess':
                    dangerous_policies.append({
                        'type': 'managed',
                        'name': policy_name,
                        'arn': policy_arn,
                        'reason': 'administratoraccess policy'
                    })
                    
        except Exception as e:
            print(f"  error checking managed policies for {username}: {e}")
        
        return dangerous_policies
    
    def check_inline_policies(self, username: str) -> List[Dict]:
        """check if user has inline policies with wildcard permissions"""
        dangerous_policies = []
        try:
            # get list of inline policy names
            response = self.iam_client.list_user_policies(UserName=username)
            policy_names = response['PolicyNames']
            
            # check each inline policy
            for policy_name in policy_names:
                policy_response = self.iam_client.get_user_policy(
                    UserName=username,
                    PolicyName=policy_name
                )
                
                # parse policy document
                import urllib.parse
                policy_document = policy_response['PolicyDocument']
                
                # check for "Action": "*"
                if self._has_wildcard_action(policy_document):
                    dangerous_policies.append({
                        'type': 'inline',
                        'name': policy_name,
                        'arn': None,
                        'reason': 'contains action wildcard'
                    })
                    
        except Exception as e:
            print(f"  error checking inline policies for {username}: {e}")
        
        return dangerous_policies
    
    def _has_wildcard_action(self, policy_document: dict) -> bool:
        """check if policy document contains action: *"""
        try:
            statements = policy_document.get('Statement', [])
            for statement in statements:
                actions = statement.get('Action', [])
                # handle both string and list formats
                if isinstance(actions, str):
                    actions = [actions]
                if '*' in actions:
                    return True
        except Exception as e:
            print(f"  warning: error parsing policy document: {e}")
        
        return False
    
    def audit_users(self):
        """scan all users and identify over-privileged ones"""
        print("\nstarting iam security audit\n")
        
        users = self.get_all_users()
        
        for username in users:
            print(f"auditing user: {username}")
            
            # check managed policies
            dangerous_managed = self.check_managed_policies(username)
            
            # check inline policies
            dangerous_inline = self.check_inline_policies(username)
            
            # combine all dangerous policies
            all_dangerous = dangerous_managed + dangerous_inline
            
            if all_dangerous:
                print(f"  flagged - user has {len(all_dangerous)} dangerous policy/policies")
                for policy in all_dangerous:
                    print(f"    - {policy['type']}: {policy['name']} ({policy['reason']})")
                
                self.flagged_users.append({
                    'username': username,
                    'dangerous_policies': all_dangerous
                })
            else:
                print(f"  user is safe")
            
            print()
        
        print(f"audit complete: {len(self.flagged_users)} over-privileged users found\n")
        
        return self.flagged_users
    
    def create_safe_policy(self):
        """create limitedaccesspolicy with s3 and ec2 readonly access"""
        policy_document = {
            "Version": "2012-10-17",
            "Statement": [
                {
                    "Effect": "Allow",
                    "Action": [
                        "s3:Get*",
                        "s3:List*",
                        "s3:Describe*",
                        "s3-object-lambda:Get*",
                        "s3-object-lambda:List*"
                    ],
                    "Resource": "*"
                },
                {
                    "Effect": "Allow",
                    "Action": [
                        "ec2:Describe*",
                        "ec2:Get*",
                        "ec2:List*"
                    ],
                    "Resource": "*"
                }
            ]
        }
        
        try:
            # try to create the policy
            response = self.iam_client.create_policy(
                PolicyName=self.safe_policy_name,
                PolicyDocument=json.dumps(policy_document),
                Description='Read-only access to S3 and EC2 for security remediation'
            )
            self.safe_policy_arn = response['Policy']['Arn']
            print(f"created new policy: {self.safe_policy_name}")
            print(f"  arn: {self.safe_policy_arn}")
            
        except self.iam_client.exceptions.EntityAlreadyExistsException:
            # policy already exists, get its arn
            print(f"policy {self.safe_policy_name} already exists")
            account_id = boto3.client('sts').get_caller_identity()['Account']
            self.safe_policy_arn = f"arn:aws:iam::{account_id}:policy/{self.safe_policy_name}"
            print(f"  using existing arn: {self.safe_policy_arn}")
            
        except Exception as e:
            print(f"error creating safe policy: {e}")
            return False
        
        return True
    
    def remediate_user(self, username: str, dangerous_policies: List[Dict]):
        """detach dangerous policies and attach safe policy"""
        print(f"\nremediating user: {username}")
        
        # detach dangerous policies
        for policy in dangerous_policies:
            try:
                if policy['type'] == 'managed':
                    # detach managed policy
                    self.iam_client.detach_user_policy(
                        UserName=username,
                        PolicyArn=policy['arn']
                    )
                    print(f"  detached managed policy: {policy['name']}")
                    
                elif policy['type'] == 'inline':
                    # delete inline policy
                    self.iam_client.delete_user_policy(
                        UserName=username,
                        PolicyName=policy['name']
                    )
                    print(f"  deleted inline policy: {policy['name']}")
                    
            except Exception as e:
                print(f"  error removing policy {policy['name']}: {e}")
        
        # attach safe policy
        try:
            self.iam_client.attach_user_policy(
                UserName=username,
                PolicyArn=self.safe_policy_arn
            )
            print(f"  attached safe policy: {self.safe_policy_name}")
            
        except Exception as e:
            print(f"  error attaching safe policy: {e}")
    
    def remediate_all(self):
        """remediate all flagged users"""
        if not self.flagged_users:
            print("\nno users to remediate!")
            return
        
        print("\nstarting automated remediation\n")
        
        # create safe policy first
        if not self.create_safe_policy():
            print("failed to create safe policy. aborting remediation.")
            return
        
        print()
        
        # remediate each flagged user
        for user_info in self.flagged_users:
            self.remediate_user(
                user_info['username'],
                user_info['dangerous_policies']
            )
        
        print(f"\nremediation complete")
        print(f"{len(self.flagged_users)} users have been secured")
        print(f"all users now have {self.safe_policy_name} attached\n")
    
    def list_user_policies(self, username: str):
        """list all policies attached to a specific user"""
        print(f"\n{username}:")
        
        # list managed policies
        try:
            response = self.iam_client.list_attached_user_policies(UserName=username)
            managed_policies = response['AttachedPolicies']
            
            if managed_policies:
                print(f"  managed policies:")
                for policy in managed_policies:
                    print(f"    - {policy['PolicyName']}")
            else:
                print(f"  managed policies: none")
                
        except Exception as e:
            print(f"  error listing managed policies: {e}")
        
        # list inline policies
        try:
            response = self.iam_client.list_user_policies(UserName=username)
            inline_policies = response['PolicyNames']
            
            if inline_policies:
                print(f"  inline policies:")
                for policy_name in inline_policies:
                    print(f"    - {policy_name}")
            else:
                print(f"  inline policies: none")
                
        except Exception as e:
            print(f"  error listing inline policies: {e}")
    
    def list_all_users_policies(self):
        """list all policies for all iam users"""
        print("\ncurrent iam user policies\n")
        
        users = self.get_all_users()
        
        if not users:
            print("no users found")
            return
        
        for username in users:
            self.list_user_policies(username)
        
        print()
    
    def run_full_audit_and_remediation(self):
        """execute complete audit and remediation workflow"""
        # step 1: audit all users
        self.audit_users()
        
        # step 2: remediate flagged users
        if self.flagged_users:
            self.remediate_all()
        else:
            print("no over-privileged users found. environment is secure!")
        
        # step 3: list all current policies
        self.list_all_users_policies()
        
        return self.flagged_users
```

```python
# execute iam security audit and remediation
auditor = IAMSecurityAuditor()
flagged_users = auditor.run_full_audit_and_remediation()
```

```python
import boto3
import time
import random
from botocore.exceptions import ClientError
from boto3.dynamodb.conditions import Key
from botocore.config import Config

class DynamoDBManager:
    def __init__(self, table_name, region="ap-south-1"):
        self.config = Config(
            retries={"max_attempts": 10, "mode": "adaptive"},
            region_name=region
        )
        self.dynamodb = boto3.resource("dynamodb", config=self.config)
        self.client = boto3.client("dynamodb", config=self.config)
        self.table = self.dynamodb.Table(table_name)

    def _execute_with_retry(self, func, *args, **kwargs):
        retries = 3
        for i in range(retries):
            try:
                return func(*args, **kwargs)
            except ClientError as e:
                code = e.response["Error"]["Code"]
                if code in [
                    "ProvisionedThroughputExceededException",
                    "ThrottlingException",
                    "RequestLimitExceeded"
                ]:
                    sleep_time = (2**i) + random.uniform(0, 1)
                    time.sleep(sleep_time)
                else:
                    raise e
        raise Exception("Max retries exceeded")

    def put_order(self, item):
        def operation():
            self.table.put_item(
                Item={**item, "Version": 1},
                ConditionExpression="attribute_not_exists(CustomerID)"
            )
            return {"status": "inserted"}
        return self._execute_with_retry(operation)

    def update_order_status(self, customer_id, order_id, new_status, version):
        def operation():
            return self.table.update_item(
                Key={"CustomerID": customer_id, "OrderID": order_id},
                UpdateExpression="SET #s = :s, Version = Version + :inc",
                ConditionExpression="Version = :v",
                ExpressionAttributeNames={"#s": "Status"},
                ExpressionAttributeValues={
                    ":s": new_status,
                    ":v": version,
                    ":inc": 1
                },
                ReturnValues="UPDATED_NEW"
            )
        return self._execute_with_retry(operation)

    def get_customer_orders(self, customer_id):
        def operation():
            items = []
            last_key = None
            while True:
                params = {
                    "KeyConditionExpression": Key("CustomerID").eq(customer_id)
                }
                if last_key:
                    params["ExclusiveStartKey"] = last_key
                response = self.table.query(**params)
                items.extend(response.get("Items", []))
                last_key = response.get("LastEvaluatedKey")
                if not last_key:
                    break
            return items
        return self._execute_with_retry(operation)

    def get_orders_by_status(self, status):
        def operation():
            response = self.table.query(
                IndexName="StatusIndex",
                KeyConditionExpression=Key("Status").eq(status)
            )
            return response.get("Items", [])
        return self._execute_with_retry(operation)

    def batch_insert_orders(self, items):
        def operation():
            with self.table.batch_writer() as batch:
                for item in items:
                    batch.put_item(Item={**item, "Version": 1})
            return {"status": "batch success"}
        return self._execute_with_retry(operation)

    def transact_orders(self, items):
        def operation():
            transact_items = []
            for item in items:
                transact_items.append({
                    "Put": {
                        "TableName": self.table.name,
                        "Item": {
                            "CustomerID": {"S": item["CustomerID"]},
                            "OrderID": {"S": item["OrderID"]},
                            "Status": {"S": item["Status"]}
                        },
                        "ConditionExpression": "attribute_not_exists(CustomerID)"
                    }
                })
            self.client.transact_write_items(TransactItems=transact_items)
            return {"status": "transaction success"}
        return self._execute_with_retry(operation)

    def delete_order(self, customer_id, order_id):
        def operation():
            self.table.delete_item(
                Key={"CustomerID": customer_id, "OrderID": order_id},
                ConditionExpression="attribute_exists(CustomerID)"
            )
            return {"status": "deleted"}
        return self._execute_with_retry(operation)
```

```python
db = DynamoDBManager("OrdersTable")

print("--- Insert Order ---")
item = {
    "CustomerID": "CUST#1",
    "OrderID": "ORD#101",
    "Status": "PENDING",
    "OrderDate": "2026-04-05",
    "TotalAmount": "500"
}
print(db.put_order(item))
```

```python
print("--- Query Orders ---")
print(db.get_customer_orders("CUST#1"))
```

```python
print("--- Query by Status (GSI) ---")
print(db.get_orders_by_status("PENDING"))
```

```python
print("--- Update Order ---")
print(db.update_order_status("CUST#1", "ORD#101", "SHIPPED", 1))
```

```python
print("--- Batch Insert ---")
batch_items = [
    {"CustomerID": "CUST#2", "OrderID": "ORD#201", "Status": "PENDING"},
    {"CustomerID": "CUST#3", "OrderID": "ORD#301", "Status": "PENDING"}
]
print(db.batch_insert_orders(batch_items))
```

```python
print("--- Transaction ---")
txn_items = [
    {"CustomerID": "CUST#4", "OrderID": "ORD#401", "Status": "CONFIRMED"},
    {"CustomerID": "CUST#5", "OrderID": "ORD#501", "Status": "CONFIRMED"}
]
print(db.transact_orders(txn_items))
```

```python
print("--- Delete ---")
print(db.delete_order("CUST#1", "ORD#101"))
```

```python
import boto3
from botocore.exceptions import ClientError
from IPython.display import display

def get_s3_details():
    s3_client = boto3.client('s3', region_name='ap-south-1')
    
    all_bucket_data = []

    try:
        # 1. Fetch all buckets
        bucket_list_response = s3_client.list_buckets()
        buckets = bucket_list_response.get('Buckets', [])

        for bucket in buckets:
            bucket_name = bucket['Name']
            
            try:
                # 2. Fetch objects for this specific bucket
                list_objects_response = s3_client.list_objects_v2(Bucket=bucket_name)
                contents = list_objects_response.get('Contents', [])
                object_keys = [obj['Key'] for obj in contents]
            except Exception:
                object_keys = ['Error or No Permission']

            all_bucket_data.append({
                'bucket': bucket_name,
                'objects': object_keys
            })
        
        return all_bucket_data

    except ClientError as e:
        return f"Authentication Error: {e}"

output = get_s3_details()
display(output)
```

```python
import json
import boto3
import uuid
from datetime import datetime

# Initialize the DynamoDB resource
dynamodb = boto3.resource('dynamodb')
table = dynamodb.Table('LabIssuesDB') 

def lambda_handler(event, context):
    try:
        # Extract the HTTP method from the API Gateway event
        method = event.get('httpMethod') or event.get('requestContext', {}).get('http', {}).get('method')
        
        # CREATE: Add a new lab issue ticket (POST)
        if method == 'POST':
            body = json.loads(event.get('body', '{}'))
            ticket_item = {
                'ticket_id': str(uuid.uuid4())[:8], # Generate an 8-character unique ID
                'timestamp': datetime.now().strftime("%Y-%m-%d %H:%M"),
                'lab_room': body.get('lab_room', 'Unknown'),
                'pc_number': body.get('pc_number', 'N/A'),
                'issue_category': body.get('issue_category', 'General'),
                'description': body.get('description', ''),
                'status': 'OPEN'
            }
            table.put_item(Item=ticket_item)
            return build_response(201, {"message": "Issue reported successfully!", "ticket": ticket_item})
            
        # READ: Fetch all open and closed tickets (GET)
        elif method == 'GET':
            response = table.scan()
            return build_response(200, {"issues": response.get('Items', [])})
            
        # UPDATE: Mark an existing ticket as closed (PUT)
        elif method == 'PUT':
            body = json.loads(event.get('body', '{}'))
            if 'ticket_id' not in body:
                return build_response(400, {"message": "Missing ticket_id for update"})
                
            table.update_item(
                Key={'ticket_id': body['ticket_id']},
                UpdateExpression="set #s = :stat",
                ExpressionAttributeNames={'#s': 'status'},
                ExpressionAttributeValues={':stat': 'CLOSED'}
            )
            return build_response(200, {"message": "Ticket closed successfully!"})
            
        # DELETE: Permanently remove a ticket from the database (DELETE)
        elif method == 'DELETE':
            body = json.loads(event.get('body', '{}'))
            if 'ticket_id' not in body:
                return build_response(400, {"message": "Missing ticket_id for deletion"})
                
            table.delete_item(Key={'ticket_id': body['ticket_id']})
            return build_response(200, {"message": "Ticket deleted successfully!"})
            
        # Fallback for unsupported methods
        return build_response(400, {"message": f"Method {method} not supported"})
        
    except Exception as e:
        return build_response(500, {"error": "Internal Server Error", "details": str(e)})

def build_response(status_code, body):
    return {
        "statusCode": status_code,
        "headers": {
            "Content-Type": "application/json",
            "Access-Control-Allow-Origin": "*", # Critical for web app CORS
            "Access-Control-Allow-Methods": "OPTIONS,POST,GET,PUT,DELETE"
        },
        "body": json.dumps(body)
    }
```
