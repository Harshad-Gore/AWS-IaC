# DynamoDB CRUD Operations with Boto3

Region: ap-south-1

```python
# create the dynamodb resource and table name
import boto3
import uuid

REGION = "ap-south-1"
dynamodb = boto3.resource("dynamodb", region_name=REGION)
table_name = f"exam-items-{uuid.uuid4().hex[:8]}"
```

```python
# create the table
table = dynamodb.create_table(
    TableName=table_name,
    KeySchema=[
        {"AttributeName": "pk", "KeyType": "HASH"},
        {"AttributeName": "sk", "KeyType": "RANGE"},
    ],
    AttributeDefinitions=[
        {"AttributeName": "pk", "AttributeType": "S"},
        {"AttributeName": "sk", "AttributeType": "S"},
    ],
    BillingMode="PAY_PER_REQUEST",
)
table.wait_until_exists()
print(table.table_status)
```

```python
# put an item
response = table.put_item(
    Item={
        "pk": "user#1",
        "sk": "profile",
        "name": "Harshad",
        "role": "student",
    }
)
print(response)
```

```python
# get and update the item
response = table.get_item(
    Key={"pk": "user#1", "sk": "profile"},
)
print(response)

response = table.update_item(
    Key={"pk": "user#1", "sk": "profile"},
    UpdateExpression="SET #r = :r",
    ExpressionAttributeNames={"#r": "role"},
    ExpressionAttributeValues={":r": "admin"},
    ReturnValues="ALL_NEW",
)
print(response)
```

```python
# delete the item and table
response = table.delete_item(
    Key={"pk": "user#1", "sk": "profile"},
)
print(response)

response = table.delete()
print(response)
```
