# S3 Operations with Boto3

Region: ap-south-1

```python
import boto3
import uuid

REGION = "ap-south-1"
DRY_RUN = True

s3 = boto3.client("s3", region_name=REGION)
```

```python
bucket_name = f"exam-s3-{uuid.uuid4().hex[:8]}"
object_key = "sample.txt"
object_body = "hello from boto3"
```

```python
if DRY_RUN:
    print(f"[DRY RUN] create_bucket: {bucket_name} in {REGION}")
else:
    s3.create_bucket(
        Bucket=bucket_name,
        CreateBucketConfiguration={"LocationConstraint": REGION}
    )
```

```python
if DRY_RUN:
    print("[DRY RUN] put_object, list_objects_v2, get_object")
else:
    s3.put_object(Bucket=bucket_name, Key=object_key, Body=object_body)
    print(s3.list_objects_v2(Bucket=bucket_name))
    response = s3.get_object(Bucket=bucket_name, Key=object_key)
    print(response["Body"].read().decode("utf-8"))
```

```python
if DRY_RUN:
    print("[DRY RUN] delete_object, delete_bucket")
else:
    s3.delete_object(Bucket=bucket_name, Key=object_key)
    s3.delete_bucket(Bucket=bucket_name)
```
