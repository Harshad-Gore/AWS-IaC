# S3 Operations with Boto3

Region: ap-south-1

```python
# create the s3 client and names
import boto3
import uuid

REGION = "ap-south-1"
client = boto3.client("s3", region_name=REGION)

bucket_name = f"exam-bucket-{uuid.uuid4().hex[:8]}"
object_key = "sample.txt"
object_body = "hello from boto3"
```

```python
# create the bucket
response = client.create_bucket(
    Bucket=bucket_name,
    CreateBucketConfiguration={"LocationConstraint": REGION},
)
print(response)
```

```python
# upload an object
response = client.put_object(
    Bucket=bucket_name,
    Key=object_key,
    Body=object_body,
)
print(response)
```

```python
# list objects and get the object
response = client.list_objects_v2(
    Bucket=bucket_name,
)
print(response)

response = client.get_object(
    Bucket=bucket_name,
    Key=object_key,
)
print(response["Body"].read().decode("utf-8"))
```

```python
# delete the object and bucket
response = client.delete_object(
    Bucket=bucket_name,
    Key=object_key,
)
print(response)

response = client.delete_bucket(
    Bucket=bucket_name,
)
print(response)
```
