# IAM Operations with Boto3

Region: ap-south-1

```python
import boto3
from botocore.exceptions import ClientError

REGION = "ap-south-1"
DRY_RUN = True

iam = boto3.client("iam", region_name=REGION)
```

```python
USER_NAME = "exam-user"
GROUP_NAME = "exam-group"
POLICY_ARN = "arn:aws:iam::aws:policy/ReadOnlyAccess"
```

```python
def maybe(action, func, **kwargs):
    if DRY_RUN:
        print(f"[DRY RUN] {action}: {kwargs}")
        return None
    try:
        return func(**kwargs)
    except ClientError as exc:
        print(f"Error during {action}: {exc}")
        return None
```

```python
maybe("create_group", iam.create_group, GroupName=GROUP_NAME)
maybe("create_user", iam.create_user, UserName=USER_NAME)
maybe("add_user_to_group", iam.add_user_to_group, GroupName=GROUP_NAME, UserName=USER_NAME)
maybe("attach_user_policy", iam.attach_user_policy, UserName=USER_NAME, PolicyArn=POLICY_ARN)
```

```python
if DRY_RUN:
    print("[DRY RUN] list_users, list_groups, list_attached_user_policies")
else:
    print(iam.list_users())
    print(iam.list_groups())
    print(iam.list_attached_user_policies(UserName=USER_NAME))
```

```python
maybe("detach_user_policy", iam.detach_user_policy, UserName=USER_NAME, PolicyArn=POLICY_ARN)
maybe("remove_user_from_group", iam.remove_user_from_group, GroupName=GROUP_NAME, UserName=USER_NAME)
maybe("delete_user", iam.delete_user, UserName=USER_NAME)
maybe("delete_group", iam.delete_group, GroupName=GROUP_NAME)
```
