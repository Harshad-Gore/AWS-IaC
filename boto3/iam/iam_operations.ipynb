# IAM Operations with Boto3

Region: ap-south-1

```python
# create the iam client and shared names
import boto3

client = boto3.client("iam")
USER_PREFIX = "exam-iam-user"
GROUP_NAME = "exam-iam-group"
POLICY_ARN = "arn:aws:iam::aws:policy/ReadOnlyAccess"
```

```python
# create iam users with a loop
for i in range(5):
    username = f"{USER_PREFIX}-{i}"
    response = client.create_user(
        UserName=username,
    )
    print(response)
```

```python
# create the iam group and attach a managed policy
response = client.create_group(
    GroupName=GROUP_NAME,
)
print(response)

response = client.attach_group_policy(
    GroupName=GROUP_NAME,
    PolicyArn=POLICY_ARN,
)
print(response)
```

```python
# add iam users to the group and attach a policy
for i in range(5):
    username = f"{USER_PREFIX}-{i}"
    response = client.add_user_to_group(
        GroupName=GROUP_NAME,
        UserName=username,
    )
    print(response)

    response = client.attach_user_policy(
        UserName=username,
        PolicyArn=POLICY_ARN,
    )
    print(response)
```

```python
# list iam users, groups, and attached policies
response = client.list_users()
print(response)

response = client.list_groups()
print(response)

response = client.list_attached_user_policies(
    UserName=f"{USER_PREFIX}-0",
)
print(response)
```

```python
# detach the policy, remove users from the group, and delete iam resources
for i in range(5):
    username = f"{USER_PREFIX}-{i}"
    response = client.detach_user_policy(
        UserName=username,
        PolicyArn=POLICY_ARN,
    )
    print(response)

    response = client.remove_user_from_group(
        GroupName=GROUP_NAME,
        UserName=username,
    )
    print(response)

    response = client.delete_user(
        UserName=username,
    )
    print(response)

response = client.delete_group(
    GroupName=GROUP_NAME,
)
print(response)
```
